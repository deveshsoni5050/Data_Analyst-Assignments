## 1. Data Cleaning & Preprocessing

Explanation
Data Cleaning is the process of handling missing values, removing unnecessary columns, and preparing the dataset for analysis.

Steps performed:
1.Load the Titanic dataset (train.csv)
2.Check missing values
3.Fill missing values in Age using mean/median
4.Fill missing values in Embarked using the most frequent value (mode)
5.Drop unnecessary columns (Name, Ticket, Cabin) for visualization

```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("train.csv")

print("First 5 rows of dataset:")
print(df.head())

# Check missing values
print("\nMissing Values in Dataset:")
print(df.isnull().sum())

# Fill missing Age values with median
df["Age"] = df["Age"].fillna(df["Age"].median())

# Fill missing Embarked values with most common value (mode)
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])

# Drop unnecessary columns
df = df.drop(columns=["Name", "Ticket", "Cabin"])

print("\nDataset after cleaning:")
print(df.isnull().sum())
```

## 2. Univariate Analysis
Explanation

Univariate analysis means analyzing a single column (variable) to understand its distribution.

We will do:

1. Age distribution (Histogram + KDE)
2. Passenger count by Sex (Bar chart)
3. Fare distribution (Boxplot)

##### (a) Age Distribution (Histogram + KDE Plot)

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.histplot(df["Age"], kde=True, bins=30)
plt.title("Distribution of Age (Histogram + KDE)")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

### Output (Expected)
-Histogram shows most passengers are between 20 to 40 years.

-KDE shows smooth distribution curve.

##### (b) Count of Passengers by Sex (Bar Chart)

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Sex", data=df)
plt.title("Count of Passengers by Sex")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.show()
```

##### (c) Distribution of Fare using Boxplot

```python
plt.figure(figsize=(8,4))
sns.boxplot(x=df["Fare"])
plt.title("Fare Distribution (Boxplot)")
plt.xlabel("Fare")
plt.show()
```

## 3. Bivariate Analysis

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="Pclass", y="Fare", data=df)
plt.title("Pclass vs Fare (Boxplot)")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
plt.show()
```

##### Output (Expected)
-1st class passengers have higher fare.

-3rd class passengers have lower fare.

##### (c) Age vs Fare Scatter Plot (Color-coded by Survived)

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="Age", y="Fare", data=df)
plt.title("Age vs Fare (Scatter Plot)")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```
